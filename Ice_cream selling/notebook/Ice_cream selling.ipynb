# Calculating Ice_cream selling data using Polynomial Linear Regression

The Data talks about the Ice_cream selling data Depending 1 attribute of Temperature (°C)


Polynomial Linear Regression General Formula:

y = a + b1*x + b2*x^2 + b3*x^3 + b4*x^4 + ....... + bN*x^N

# 1 - first step import library

import model LinearRegression , PolynomialFeatures and pandas to read csv file and matplotlib for visualization

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd
import matplotlib.pyplot as plt
```

## 2 - Read File

```python
dataset = pd.read_csv(r"D:\Projects\projects\Machine-learning-algorithms\Ice_cream selling\data\Ice_cream selling data.csv",sep = ",")
```

## 3 - Data analysis

info to know more information about Data

## 4 - visualization

```python
plt.scatter(dataset['Temperature (°C)'],dataset['Ice Cream Sales (units)'])
plt.xlabel('Temperature (°C)')
plt.ylabel('Ice Cream Sales (units)')
plt.show()
```

## 5 - Training Model

```python
reg = LinearRegression()
reg.fit(dataset[['Temperature (°C)']],dataset['Ice Cream Sales (units)'])
```

```python
plt.xlabel('Temperature (°C))')
plt.ylabel('Ice Cream Sales (units)')
plt.scatter(dataset['Temperature (°C)'],dataset['Ice Cream Sales (units)'])   
original = dataset['Temperature (°C)']
prediction = reg.predict(dataset[['Temperature (°C)']])
plt.plot(original, prediction,color= "green", label = "linear regression")
plt.legend()
plt.show()
```

## Very bad model prediction, so: It is not correct to apply linear regression for this dataset

## I will Try polynomial regression, one of the regression types

We call the PolynomialFeatures function to create a polynomial regression object
We specify the degree (N) of the polynomial when calling this function

```python
polynomial_regression = PolynomialFeatures(degree = 4)
x_polynomial = polynomial_regression.fit_transform(dataset[['Temperature (°C)']])
```

## Training Polynomial Model 

```python
reg = LinearRegression()
reg.fit(x_polynomial,dataset['Ice Cream Sales (units)'])
```

## visualization

```python
y_head = reg.predict(x_polynomial)
plt.plot(dataset['Temperature (°C)'],y_head,color= "red",label = "polynomial regression")
plt.legend()
plt.scatter(dataset['Temperature (°C)'],dataset['Ice Cream Sales (units)'])   
plt.show()
```

its definitely fits, polynomial regression is the right choice
Now let's make N=3 or 4 and see if we increase the polynomial degree, will it fit better?
